#### Load the image and resize it to a manageable size. You may want to convert it to grayscale.

```python
dir_to_pics = pathlib.Path("homework images vol 1")
pictures = list(dir_to_pics.glob("*.jpg"))
pic = pictures[6]

im = skimage.io.imread(pic)
im = skimage.img_as_float(skimage.color.rgb2gray(im))
im = skimage.transform.rescale(im, scale=0.2, multichannel=False)
fig,ax = plt.subplots(figsize=(15,15))
ax.imshow(im, cmap="gray", vmin=0, vmax=1)
```

#### Look for edges using the Canny edge detector.

```python
canny_filter = skimage.feature.canny(im, sigma = 3, low_threshold=0.01, high_threshold=0.050)
fig,ax = plt.subplots(figsize=(15,15))
ax.imshow(canny_filter, cmap="gray", vmin=0, vmax=1)
```

#### Look for four long straight lines using the Hough line detector.

```python
#https://scikit-image.org/docs/dev/api/skimage.transform.html#skimage.transform.hough_line_peaks
hspace, angles, dists = skimage.transform.hough_line(canny_filter)
hspace, angles, dists = skimage.transform.hough_line_peaks(hspace, angles, dists, num_peaks=4)
```

#### Get the 4 lines and intersections

```python
def drawline(ax, angle, dist):
    #from exercise 3
    ## Now we want to draw the line in image space
    # This is one point on the line
    p1 = np.array([dist * np.cos(angle), dist * np.sin(angle)])
    # This is the unit vector pointing in the direction of the line (remember what theta means in Hough space!)
    linedir = np.array([np.cos(angle + np.pi / 2), np.sin(angle + np.pi / 2)])
    # These are two points very far away in two opposite directions along the line
    p0 = p1 - linedir * 1000
    p2 = p1 + linedir * 1000
    # We now draw a line through p0 and p2, without rescaling the axes.
    ax.plot([p0[0], p2[0]], [p0[1], p2[1]], scalex=False, scaley=False, color="black")
    return np.cross(np.array([p0[0],p0[1],1]),np.array([p2[0],p2[1],1]))

def distance_to_center(p):
    center = np.array([im.shape[1]/2, im.shape[0]/2])
    return np.linalg.norm(p-center)

fig,ax = plt.subplots(figsize=(15,15))
ax.imshow(im, cmap="gray", vmin=0, vmax=1)
lines = []
intersections =[]

for angle,dist in zip(angles,dists):
    lines.append(drawline(ax, angle, dist))

#get the intersection between two lines
for i in range(0,len(lines)-1):
    for j in range(1,len(lines)):
        intersection = np.cross(lines[i],lines[j])
        if not intersection.all()==0:
            intersections.append([intersection[0]/intersection[2],intersection[1]/intersection[2]])
            print("Cartesian coordinates for point intersection: ", intersection[0]/intersection[2], intersection[1]/intersection[2])
            ax.plot(intersection[0]/intersection[2], intersection[1]/intersection[2], 'bx')
```

##### The intersections closest to the center will be kept (Total 4 intersections)

```python
intersections = sorted(intersections, key=distance_to_center)
intersections = intersections[:4]
intersections = np.asarray(intersections)
len(intersections)
```

```python
#order the intersections to lower left, lower right, uper right, upper left
cornerpoints=[]
lower_points = sorted(intersections, key=lambda x: x[1], reverse=True)[:2]
uppper_points = sorted(intersections, key=lambda x: x[1], reverse=True)[2:]
cornerpoints.append(sorted(lower_points, key=lambda x: x[0], reverse=False)[0])
cornerpoints.append(sorted(lower_points, key=lambda x: x[0], reverse=False)[1])
cornerpoints.append(sorted(uppper_points, key=lambda x: x[0], reverse=True)[0])
cornerpoints.append(sorted(uppper_points, key=lambda x: x[0], reverse=True)[1])
```

#### Straightening an image distorted by perspective (rectification)

```python
w,h = 880,1250 # we assume 10 pixels per mm
A = np.array([0,h])
B = np.array([w,h])
C = np.array([w,0])
D = np.array([0,0])

invtf = skimage.transform.estimate_transform("projective",
    src=np.vstack((A,B,C,D)),
    dst=np.vstack((cornerpoints)))

tim = skimage.transform.warp(image=im, 
                             inverse_map=invtf, 
                             output_shape=(h,w),
                             mode="edge"
)
fig,ax = plt.subplots(figsize=(10,10))
ax.imshow(tim, cmap="gray", vmin=0, vmax=1)
```

#### Cut a square bounding box around each connected component and resize that to 28x28

```python
#Get numbers from images

px_per_mm = 2

bg = skimage.morphology.dilation(tim, skimage.morphology.disk(5 * px_per_mm))
tim2 = tim / bg
tim3 = tim2 < 0.90
tim4 = skimage.morphology.dilation(tim3, skimage.morphology.disk(2 * px_per_mm))


regions = skimage.measure.regionprops(skimage.morphology.label(tim4))

def preprocess(digit):
    digit = 1.0-digit
    digit = digit/np.max(digit)
    digit = np.clip(digit, a_min=0.0, a_max=1.0)
    return digit

#find a region with a number
def goodregion(r):
    w = (r.bbox[3] - r.bbox[1] + 1) / px_per_mm
    h = (r.bbox[2] - r.bbox[0] + 1) / px_per_mm
    if w < 5 or w > 120:
        return False
    if h < 5 or h > 120:
        return False
    if w / h > 4 or w / h < 1/4:
        return False
    return True

#crop the numbers from the image
digits = []
for r in regions:
    if goodregion(r):
        digit = tim2[r.bbox[0]:r.bbox[2],r.bbox[1]:r.bbox[3]]
        dim = max(digit.shape)
        sq_digit = np.ones((dim,dim))
        row0 = (dim-digit.shape[0])//2
        col0 = (dim-digit.shape[1])//2
        sq_digit[row0:(row0+digit.shape[0]),col0:(col0+digit.shape[1])] = digit
        sq_digit = skimage.transform.resize(sq_digit, (28,28))
        digits.append(sq_digit)
        fig, ax = plt.subplots(figsize=(3, 3))
        ax.imshow(sq_digit, cmap="gray", vmin=0, vmax=1)
        ax.set(xticks=[], yticks=[])
        plt.show()
        
norm_digits=[]
for i, digit in enumerate(digits):
    digitp = preprocess(digit)
    norm_digits.append(digitp)
    fig, ax = plt.subplots(figsize=(3, 3))
    ax.imshow(digitp, cmap="gray")
```
